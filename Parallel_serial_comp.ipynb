```python
from multiprocessing import Pool
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import graphviz
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix 
import time
from itertools import product
```

```python
disease_data = pd.read_csv('normalizedcardio.csv', sep = ',', header = None)
```

```python
disease_data
```

```python
def dectree(disease_data):
    X = disease_data.values[:, 1:12] 
    Y = disease_data.values[:, 12]
    X_train, X_test, y_train, y_test = train_test_split(  
    X, Y, test_size = 0.3, random_state = 100)  
    clf_gini = DecisionTreeClassifier(criterion = "gini", 
    random_state = 100,max_depth=3, min_samples_leaf=5)
    clf_gini.fit(X_train, y_train)
    y_pred = clf_gini.predict(X_test) 
    acc = accuracy_score(y_test,y_pred)*100
    tree = clf_gini.fit(X_train, y_train)
    export_graphviz(tree, out_file = "mytree.dot")
    with open("mytree.dot") as f:
        dot_graph = f.read()
    graph = graphviz.Source(dot_graph)  
    graph.view()
    #print("Accuracy: ",acc)
    return disease_data
```

```python
def parallelize_dataframe(df, func, n_cores=4):
    df_split = np.array_split(df, n_cores)
    pool = Pool(n_cores)
    df = pd.concat(pool.map(func, df_split))
    pool.close()
    pool.join()
    return df
def parallel_run():
    paralleldf = parallelize_dataframe(disease_data, dectree)
```

```python
def serial_run():
    serialdf = dectree(disease_data)
```

```python
start_serial = time.time()
for i in range(4):
    serial_run()
end_serial = time.time()
print("Time taken for serial computation : ",end_serial - start_serial, " seconds")
print()
start_para = time.time()
parallel_run()
end_para = time.time()
print("Time taken for parallel computation : ",end_para - start_para, " seconds")
print()
```
